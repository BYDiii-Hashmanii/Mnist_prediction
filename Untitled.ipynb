```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score , roc_curve, auc, f1_score,confusion_matrix, ConfusionMatrixDisplay
```

```python
# data loading 

mnist= fetch_openml('mnist_784')

x= mnist['data'].to_numpy()
y = mnist['target'].astype(int)
print("Hi")
```

```python
# Checking Shape of X and Y Values
print(x.shape)
print(y.shape)
print("Shape")
```

```python
# splitting the data 
x_train,x_test = x[:60000],x[60000:]
y_train,y_test= y[:60000],y[60000:]
```

```python
# picking up a random value 
some_digit = x[3600]
# convert it into 28 , 28 image
some_digit_image=some_digit.reshape(28,28)

plt.imshow(some_digit_image,cmap='gray')

plt.show()

print(y[3600])
```

```python
# Training of Model
clf= LogisticRegression(tol=0.1,max_iter=2000)
clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)
```

```python
score = accuracy_score(y_pred,y_test)
print(score*100)
```

```python
# Checking F1 score

# Calculate F1 score for multiclass
f1_macro = f1_score(y_test, y_pred, average='macro')
f1_weighted = f1_score(y_test, y_pred, average='weighted')

print("F1 Score (macro):", f1_macro)
print("F1 Score (weighted):", f1_weighted)
```

```python
# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.grid(False)
plt.show()
```

```python
# Drawing ROC Curve
# Predict probabilities
y_score = clf.predict_proba(x_test)

# One-vs-Rest: Binarize the labels for class "5"
y_test_binary = (y_test == 5).astype(int)
y_score_binary = y_score[:, 5]  # Probabilities for class "5"

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test_binary, y_score_binary)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for class 5')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
